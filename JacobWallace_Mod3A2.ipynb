```python
## Imports
import numpy as np 
import matplotlib.pyplot as plt 
import cv2
import glob
```

```python
# Get Images
imgs = glob.glob('img/*.jpg')
print(imgs)
```

```python
# Display source images
fig = plt.figure(figsize=(10, 10))
rows = 2
columns = 2
for i,filename in enumerate(imgs):
    img = cv2.imread(filename)
    fig.add_subplot(rows, columns, i+1)
    plt.imshow(img)
```

```python
# Normalize and Resize images
dsize = (425, 425)
fig = plt.figure(figsize=(10, 10))
rows = 2
columns = 2

for i,filename in enumerate(imgs):
    src_img = cv2.imread(filename)
    fig.add_subplot(rows, columns, i+1)
    img_resize = cv2.resize(src_img, dsize, interpolation = cv2.INTER_AREA)
    img_norm = np.zeros(dsize)
    img_final = cv2.normalize(img_resize, img_norm, 0, 100, cv2.NORM_MINMAX)
    plt.imshow(img_final)
```
